```python
import pandas as pd
import numpy as np
import os
```

```python
manual = pd.read_csv("manual_clean.csv")
manual
```

```python
for i, row in manual.iterrows():
    if not pd.isnull(row['Alt']):
        try:
            os.replace(f"data/train/{row['Class']}/{row['File']}", f"data/train/{row['Alt']}/{row['File']}")
        except:
            continue
    else:
        try:
            os.remove(f"data/train/{row['Class']}/{row['File']}")
        except:
            continue
```
